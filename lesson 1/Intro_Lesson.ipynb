# Найдем частное двух чисел
* Начало
* Введите два числа
* Введите первое число
* Введите второе число
* Если второе число = 0
* Вывод: решений нет
* Иначе: первое число / второе число = частное
* Вывод: частное
* Конец


```python
print(f"Введите два числа")
a = int(input('Введите первое число: '))
b = int(input('Введите второе число: '))
if b == 0:
    print("Решений нет")
else:
    c = a/b
    print(f'Частное двух чисел = {c}')
```

# Найдем сумму и произведение цифр трехзначного числа
* Начало
* Вывод: введите трехзначное число
* Если число трехзначное - продолжаем
* Если число не трехзначное - возвращаемся к первому выводу
* С трехзначным числом: определяем первую цифру: ххх // 100 = 1 цифра
* Опеделяем вторую цифру: (xxx - 1 цифра * 100) // 10 = 2 цифра
* Определяем третью цифру: (xxx - 1 цифра - 2 цифра) = 3 цифра
* Сумма = 1 цифра  + 2 цифра + 3 цифра
* Произведение = 1 цифра * 2 цифра * 3 цифра
* Вывод: сумма и произведение
* Конец

```python
x = int(input("Введите трехзначное число: "))
while len(str(x)) !=3:
    print ("Вы ошиблись. Это не трехзначное число")
    x = int(input())
if len(str(x)) == 3:
    a = x // 100
    b = (x-a*100) // 10 # = x % 100 // 10
    c = x - a*100 - b*10 # = x % 10
    summa = a+b+c
    multi = a*b*c
    print(f'Сумма цифр трехзначного числа = {summa}, а произведение цифр = {multi}.')
```

# Вычислим значение y при заданном значении x с условием расчета
##### если x = 0, y = 0; x > 0, y = 2x - 10; x < 0, y = 2|x|-1
* Начало
* Введите значение x
* Проверка значения x: если x > 0, вычисляем y = 2x - 10
* Если нет, то проверяем, что x < 0, вычисляем y = 2|x|-1
* Если нет, то x = 0, значит y = 0
* Выводим значение y
* Конец

```python
x = float(input('Введите значение x: '))
if x>0:
    y = 2*x-10
elif x<0:
    y = 2*abs(x)-1
else:
    y = 0
print(f'Значение y = {y}')    
```

# Сравним 3 числа и найдем максимальное (2 способа)
### 1 способ
* Начало
* Введите числа: первое, второе и третье
* Если первое число больше второго и третьего, первое число - максимальное
* Если нет, то если второе число больше первого и третьего, значит, второе число - максимальное
* Если нет, то третье число максимальное
* Вывод максимального числа
* Конец

### 2 способ
* Начало
* Введите числа: первое, второе и третье
* Примем, что Максимальное число = первое, и первое число = m.
* Если m < второго числа, принимаем за m второе число
* Если нет, то m остается равным первому числу;
* Если m < третьего числа, принимаем за m третье число;
* Если же нет, то m остается равным первому числу.
* Вывод m
* Конец

```python
# 1 способ
a = float (input('Введите первое число: '))
c = float (input('Введите второе число: '))
b = float (input('Введите третье число: '))

if (a>b) & (a>c):
    print(f'Максимальное число = {a}')
elif (b>a) & (b>c):
    print(f'Максимальное число = {b}')
else:
    print(f'Максимальное число = {c}')   
```

```python
# 2 способ
a = float (input('Введите первое число: '))
c = float (input('Введите второе число: '))
b = float (input('Введите третье число: '))

m = a
if m<b:
    m=b
if m<c:
    m=c
print(f'Максимальное число = {m}')    
```

# Перевод байтов в килобайты, и наоборот
* Начало
* Ввод числа
* Ввод направления перевода
* Если направление перевода = k, делим число на 1024
* Если направление перевода = b, умножаем число на 1024
* Если направление перевода указано неверно, сообщаем об ошибке
* Выводим результат на экран
* Конец

```python
x = int(input('Введите число для преобразования в байты или килобайты: '))
direction = str(input('Выберите направление перевода: k = килобайты, b = байты: '))

if direction == 'k':
    y = x/1024
else:
    y = x * 1024
    
if direction != 'k' and direction != 'b':
    print ('Неверный ввод.')
else:
    print (f'{x} = {y}{direction}')
```
