```python
#Выполнить логические побитовые операции «И», «ИЛИ» и др. над числами 5 и 6. Выполнить над числом 5 побитовый сдвиг вправо и влево на два знака.

result1 = 5 & 6 # операция И

result2 = 5 | 6 # операция ИЛИ

result3 = 5 ^ 6 # операция ИСКЛЮЧИТЕЛЬНОЕ ИЛИ

result4 = str(~5)+', '+str(~6) # унарный оператор

result5 = 5 << 2 # сдвиг влево

result6 = 5 >> 2 # сдвиг вправо

print (
f'Результат логической операции И для чисел 5 и 6: {result1}\n\
Результат логической операции ИЛИ для чисел 5 и 6: {result2}\n\
Результат логической операции Исключительного ИЛИ для чисел 5 и 6: {result3}\n\
Результат работы унарного оператора с числами 5 и 6: {result4}, соответственно.\n\
Результат побитового сдвига влево числа 5: {result5}\n\
Результат побитового сдвига вправо числа 5: {result6}'
)
```
