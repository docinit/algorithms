```python
#1. В диапазоне натуральных чисел от 2 до 99 определить,
#сколько из них кратны каждому из чисел в диапазоне от 2 до 9.
list1=[i for i in range(2,100)]
list2=[i for i in range(2,10)]
#для индекса
i=0
#для формирования таблицы
counter=[0]*len(list2)
#проверяем каждое число из list2 (от 2 до 9)
for num2 in list2:
    #простой счетчик
    count = 0
    #проверяем каждое число из list1 (от 2 до 99)
    for num1 in list1:
        #если вторые делятся на первое без остатка, то
        if num1%num2==0:
            # увеличиваем счетчик на 1
            count+=1
    counter[i]=count
    i+=1
    print(f'Количество чисел в диапазоне от 2 до 99, которые делятся без остатка на {num2}, равно {count}.')

print('\n','Или вкратце: ','\n')    
for numerus,values in enumerate(list2):
        print(list2[numerus],':',counter[numerus])
```

```python
# В массиве случайных целых чисел поменять местами минимальный и максимальный элементы.

# Алгоритм (псевдокод)
# начало
# создать пустой список
# ввести значения
# проверить введенные значения
# вывод сообщений об ошибке
# проверить числа на условие a>b
# создать начало отчета для минимального (min_num) и максимального (max_num) чисел:
        # max_num = верхний предел диапазона + 1, max_num = нижний предел -1
# заполнить пустой список
# если k-элемент списка > max_num, то max_num=k-элемент
# imax=k
# если k-элемнт списка < min_num, то min_num = k-элемент
# imin=k
# вывод позиций максимального и минимального элементов
# вывод начального списка
# обмен значений: listl[imax],listl[imin]=listl[imin],listl[imax]
# вывода результирующего списка
# конец



import random
listl=[]
while True:
    try:
        a=int(input('Введите минимальное число списка: '))
        b=int(input('Введите максимальное число списка: '))
        n=int(input('Введите количество элементов списка: '))        
        if a<b:
            min_num=b+1
            max_num = a-1            
            for i in range(0,n):
                listl.append(random.randint(a,b))
            for k,i in enumerate(listl):
                if k<=len(listl)-1:
                    if listl[k]>max_num:
                        max_num=listl[k]
                        imax=k
                    elif listl[k]<min_num:
                        min_num=listl[k]
                        imin=k
            print (f'Максимальное значение {max_num} находится на {imax+1} месте, а минимальное {min_num} - на {imin+1} месте.')            
            print(f'До смены мест значений массив выглядит так: {listl}')
            listl[imax],listl[imin]=listl[imin],listl[imax]
            print(f'После смены мест значений массив выглядит так: {listl}')
            break
        else:
            print ('Вы ввели неверные значения. Попробуйте ввести правильные значения еще раз.') 
    except ValueError:
        print ('Вы ввели неверные значения. Попробуйте ввести правильные значения еще раз.')
```

```python
# 4. Определить, какое число в массиве встречается чаще всего.

# Алгоритм (псевдокод)
# начало
# создать пустой список для первого и второго списка
# ввести значения: минимум и максимум диапазона, количество чисел
# вывод сообщений об ошибках
# проверить введенные значения: минимум < максимум, количество > (максимум-минимум)
# заполнить пустой список
# для каждого элемент в списке проверять условие:

# для каждого элемента в списке выполнить: увеличить значение m, если какой-нибудь элемент равен этому элементу.
# если m>p: добавить элемент в список максимальных
# m=p
# если m>p, m=p
# вывод позиций максимального и минимального элементов
# вывод начального списка
# обмен значений: listl[imax],listl[imin]=listl[imin],listl[imax]
# вывода результирующего списка
# конец

import random
listl=[]
max_list=[]
m,p=0,0
while True:
    try:
        a=int(input('Введите минимальное число списка: '))
        b=int(input('Введите максимальное число списка: '))
        n=int(input('Введите количество элементов списка: '))        
        if a<b and n>(b-a):
            for i in range(0,n):
                listl.append(random.randint(a,b))
            for x in listl:
                for i in listl:
                    if i == x:
                        m+=1
                if m>p:
                    max_list=[]
                    max_list.append(x)
                    p=m                        
                elif m==p:
                    max_list.append(x)
                    p=m
                m=0
            max_list=list(set(max_list))
            print(f'Чаще всех в списке встречается число (-а): ')
            print(*max_list, sep = ', ')            
            print('Изначальный список чисел:\n', listl)
            break
        else:
            print ('Вы ввели неверные значения. Попробуйте ввести правильные значения еще раз.') 
    except ValueError:
        print ('Вы ввели неверные значения. Попробуйте ввести правильные значения еще раз.')
```

```python
#6. В одномерном массиве найти сумму элементов, находящихся между минимальным и максимальным элементами.
#Сами минимальный и максимальный элементы в сумму не включать.


import random
listl=[]
mult1,mult2=0,0
while True:
    try:
        a=int(input('Введите минимальное число списка: '))
        b=int(input('Введите максимальное число списка: '))
        n=int(input('Введите количество элементов списка: '))        
        if a<b:
            min_num=b+1
            max_num = a-1            
            for i in range(0,n):
                listl.append(random.randint(a,b))
            for k,i in enumerate(listl):
                if k<=len(listl)-1:
                    if listl[k]>max_num:
                        max_num=listl[k]
                        imax=k
                    elif listl[k]<min_num:
                        min_num=listl[k]
                        imin=k
            if imin<imax:
                listl2=listl[imin+1:imax]
            else:
                listl2=listl[imax+1:imin]
            print (f'Максимальное значение = {max_num}, а минимальное = {min_num}.')
            print (f'Их индексы (первые вхождения): {imax}, {imin}, соответственно.')            
            print(f'Сумма значений между позициями этих элементов = {sum(listl2)}')
            print('\nПросуммированные значения: ',listl2, '\nЗначения списка: ',listl)
            break
        else:
            print ('Вы ввели неверные значения. Попробуйте ввести правильные значения еще раз.') 
    except ValueError:
        print ('Вы ввели неверные значения. Попробуйте ввести правильные значения еще раз.')
```
